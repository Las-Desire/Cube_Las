```python
import keras
import numpy as np
import matplotlib.pyplot as pp
```

```python
x_shape=(500)
X=np.random.random(x_shape)
#평균
mean=np.random.randint(-100,100)
#표준정규
std=70

X=np.random.normal(mean,std,x_shape)

X=np.random.random(x_shape)*500-255
pp.plot(np.arange(0,500,1),X,"ro")
```

```python
def random():
    return np.random.randint(-100,100)*np.random.random(1)
a=random()
b=random()
c=random()

def f(x):
    return a*x*x+b*x+c
Y=f(X)
print(a,b)
pp.plot(X,Y,"ro")
```

```python
model=keras.models.Sequential()
model.add(layer=keras.layers.Dense(10,input_shape=(1,)))
model.add(layer=keras.layers.Dense(10))
model.add(layer=keras.layers.Dense(200))
model.add(layer=keras.layers.Dense(100))
model.add(layer=keras.layers.Dense(200))
model.add(layer=keras.layers.Dense(100))
model.add(layer=keras.layers.Dense(200))
model.add(layer=keras.layers.Dense(100))
model.add(layer=keras.layers.Dense(200))
model.add(layer=keras.layers.Dense(100))
model.add(layer=keras.layers.Dense(200))
model.add(layer=keras.layers.Dense(1))
```

```python
model.compile(loss="mse",optimizer='adam')
model.fit(X,Y,epochs=300000)
```

```python
Y_=model.predict(X)
pp.plot(X,Y_,"ro",color='b')
pp.plot(X,Y,"ro")
```
