```python
import keras
import numpy as np
import matplotlib.pyplot as pp
```

```python
x_shape=(50)
X=np.random.random(x_shape)
#평균
mean=np.random.randint(-100,100)
#표준정규
std=70

X=np.random.normal(mean,std,x_shape)

X=np.random.random(x_shape)*500-255
pp.plot(np.arange(0,50,1),X,"ro")
```

```python
def random():
    return np.random.randint(-100,100)*np.random.random(1)
a=random()
b=random()
c=random()
d=random()

def f(x):
    return (a*x*3+1)*(b*x*2+10)*(c*x*2+2)
Y=f(X)
print(a,b)
pp.plot(X,Y,"ro")
```

```python
model=keras.models.Sequential()
model.add(layer=keras.layers.Dense(10,input_shape=(1,)))
model.add(layer=keras.layers.Dense(20,activation='relu'))
model.add(layer=keras.layers.Dense(40,activation='relu'))
model.add(layer=keras.layers.Dense(80,activation='relu'))
model.add(layer=keras.layers.Dense(40,activation='relu'))
model.add(layer=keras.layers.Dense(20,activation='relu'))
model.add(layer=keras.layers.Dense(10,activation='relu'))
model.add(layer=keras.layers.Dense(1))
```

```python
model.compile(loss=keras.losses.mean_absolute_error,optimizer='adam')
model.fit(X,Y,epochs=100000)
```

```python
Y_=model.predict(X)
pp.plot(X,Y_,"ro")
```
